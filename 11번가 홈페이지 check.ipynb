```python
import pandas as pd
product_df = pd.read_csv("product_link_complete.csv", encoding='cp949',index_col=0)

product_df
```

```python
product_df["link"]
```

```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import random
```

```python
for i in range(len(product_df["link"])):
    url = product_df["link"][i]
    
    driver = webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    driver.implicitly_wait(3)
    
    iframe = driver.find_element(By.ID, "ifrmReview") 
    driver.switch_to.frame(iframe)
    
    driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[1]/button").send_keys(Keys.ENTER)
    time.sleep(random.uniform(0,2))
    
    driver.quit()
    
    print(i,"번 link 확인")
```
