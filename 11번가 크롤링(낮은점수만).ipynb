```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import random

import pandas as pd
```

```python
url = 'https://www.11st.co.kr/products/3726844068?inpu=&trTypeCd=22&trCtgrNo=895019'

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

# 평점 고르는 버튼 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[1]/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

# 평점 1점 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[2]/dl/dd[6]/label/span").click()
time.sleep(random.uniform(0,2))

# 더보기 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))


date_lst = []
name_lst = []
grade_lst = []
rev_lst = []
kkuk_lst = []
    
date1 = driver.find_elements(By.CLASS_NAME, "date")
for i in date1:
    date_lst.append(i.text)
    
name1 = driver.find_elements(By.CLASS_NAME, "name")
for i in name1:
    name_lst.append(i.text)
    
grade1 = driver.find_elements(By.CSS_SELECTOR,'p.grade > span > em')
for i in grade1:
    grade_lst.append(i.text)
        
rev1 = driver.find_elements(By.CLASS_NAME,'cont_review_hide.text-expanded')
for i in rev1:
    rev_lst.append((i.text).replace('\n',''))   
        
kkuk1 = driver.find_elements(By.CLASS_NAME,'kkukCount')
for i in kkuk1:
    kkuk_lst.append((i.text))

# 평점 고르는 버튼 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[1]/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

# 평점 2점 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[2]/dl/dd[5]/label/span").click()
time.sleep(random.uniform(0,2))

# 더보기 누르기
# driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
# time.sleep(random.uniform(0,2))

date2 = driver.find_elements(By.CLASS_NAME, "date")
for i in date2:
    date_lst.append(i.text)
    
name2 = driver.find_elements(By.CLASS_NAME, "name")
for i in name2:
    name_lst.append(i.text)
    
grade2 = driver.find_elements(By.CSS_SELECTOR,'p.grade > span > em')
for i in grade2:
    grade_lst.append(i.text)
        
rev2 = driver.find_elements(By.CLASS_NAME,'cont_review_hide.text-expanded')
for i in rev2:
    rev_lst.append((i.text).replace('\n',''))   
        
kkuk2 = driver.find_elements(By.CLASS_NAME,'kkukCount')
for i in kkuk2:
    kkuk_lst.append((i.text))

        
data = pd.DataFrame([ x for x in zip(date_lst,name_lst,grade_lst,rev_lst,kkuk_lst)],columns = ['날짜','ID','평점','리뷰','꾹'])    
print(data)

```
