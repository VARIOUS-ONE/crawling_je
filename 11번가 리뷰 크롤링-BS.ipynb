```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.11st.co.kr/products/2217098035?inpu=&trTypeCd=22&trCtgrNo=895019'

data = requests.get(url)

soup = bs(data.text,'html.parser')
```

```python
Product_Name = soup.select('title')
Product_Name
```

```python
reviews = soup.select('span.cont')
review = []
for r in reviews:
    review.append(r.text)

rev = []
for i in review :
    rev.append(i.strip())

rev
        
```
