### 날짜

```python
url = product_df["1"][0]

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

date_lst = []
date = driver.find_elements(By.CLASS_NAME, "date")
for i in date:
    date_lst.append(i.text)


print(date_lst)
```

### 이름

```python
url = product_df["1"][0]

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

name_lst = []
name = driver.find_elements(By.CLASS_NAME, "name")
for i in name:
    name_lst.append(i.text)

    
driver.quit()    
print(name_lst)
```

### 평점

```python
url = product_df["1"][0]

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

grade_lst = []
grade = driver.find_elements(By.CSS_SELECTOR,'p.grade > span > em')
for i in grade:
    grade_lst.append(i.text)

driver.quit()    
print(grade_lst)
```

### 리뷰

```python
url = product_df["1"][0]

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))


rev_lst = []
rev = driver.find_elements(By.CLASS_NAME,'cont_review_hide.text-expanded')
for i in rev:
    rev_lst.append((i.text).replace('\n',''))
    


print(rev_lst)
```

### 꾹

```python
url = product_df["1"][0]

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

iframe = driver.find_element(By.ID, "ifrmReview") # id가 mainFrame이라는 요소를 찾아내고 -> iframe임
driver.switch_to.frame(iframe) # 이 iframe이라는 요소로 focus한다.

driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))


kkuk_lst = []
kkuk = driver.find_elements(By.CLASS_NAME,'kkukCount')
for i in kkuk:
    kkuk_lst.append((i.text))
    


print(kkuk_lst)
```
