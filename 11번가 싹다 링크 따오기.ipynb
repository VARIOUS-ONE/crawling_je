```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import random

import pandas as pd
```

```python
driver = webdriver.Chrome("chromedriver.exe")
url1 = 'https://www.11st.co.kr/disp/DTAction.tmall?method=&viewType=L&pageNum=1&pageRows=10&pageSize=100&kwd=&researchFlag=false&researchChkFlag=N&sortType=&sortCd=A&prdTab=T&lCtgrNo=0&mCtgrNo=0&sCtgrNo=0&dCtgrNo=0&dispCtgrNo=70249&dispCtgrCd=+++&xzone=&minPrice=0&maxPrice=0&selMthdCd=&custBenefit=&dlvType=&buySatisfy=&goodsType=&martPrdKwd=&kwdInCondition=&exceptKwdInCondition=&researchFlag=false&ctgrNo=70249&srCtgrNo=70249&clearAll=&schFrom=&referMethod=&dispCtgrNoList=&naviAnchor=&ctgrType=A&currCtgrTyp=&startDt=&endDt=&brandCdArr=#navi'
driver.get(url1)
# 드라이버 열릴 때까지 최대 3초 기다리기
driver.implicitly_wait(3)

# xpath가 순서대로 나와서 f string 으로 변수에 저장함

product_name = []
link = []

for i in range(100):
    # 상품 이름과 링크의 xpath
    # for문과 f string이용
    name_x = f'/html/body/div[2]/div[2]/div/div[3]/div[3]/ul/li[{i+1}]/div[1]/div/div[3]/div[1]/a'
    link_x = f'/html/body/div[2]/div[2]/div/div[3]/div[3]/ul/li[{i+1}]/div[1]/div/div[2]/div[1]/a'

    temp1 = driver.find_element(By.XPATH, name_x).text
    temp2 = driver.find_element(By.XPATH, link_x).get_attribute('href')
    
    product_name.append(temp1)
    link.append(temp2)
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    if i % 10 == 0:
        print(i)
    
    time.sleep(random.uniform(0,2))

driver.quit()

df1 = pd.DataFrame([ x for x in zip(product_name, link)],columns = ['상품명','링크'])
df1.to_csv("생건.csv", encoding = "utf-8-sig")
print("url1 완료")
















```
